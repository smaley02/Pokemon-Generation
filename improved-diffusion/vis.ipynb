```python
import numpy as np
```

```python
iteration = "1710000"
samples = 900
# directory = f"latest_samples/ddim"
```

```python
# npz_data = np.load(f'log_256/{directory}/samples_{str(samples)}x256x256x3.npz')
npz_data = np.load(f'log_256/samples_{str(samples)}x256x256x3.npz')
```

```python
image_data = npz_data["arr_0"]
```

```python
image_data.shape
```

```python
from PIL import Image
import os
ind = 0

img_dir = f"log_256/images_{iteration}_{samples}"

os.makedirs(img_dir, exist_ok=True)

for np_arr in image_data:
    image = Image.fromarray(np_arr)
    # path = f"log_256/{directory}/images/sample_{ind}_{iteration}.png"
    path = f"{img_dir}/sample_{ind}_{iteration}.png"
    image.save(path)
    ind += 1
print(f"saved {ind+1} samples")
```

```python
def create_collage(width, height, listofimages):
    cols = 30
    rows = 30
    if samples == 32:
        cols = 5
        rows = 5
    thumbnail_width = width//cols
    thumbnail_height = height//rows
    size = thumbnail_width, thumbnail_height
    new_im = Image.new('RGB', (width, height))
    ims = []
    for p in listofimages:
        im = Image.open(p)
        im.thumbnail(size)
        ims.append(im)
    i = 0
    x = 0
    y = 0
    for col in range(cols):
        for row in range(rows):
            print(i, x, y)
            new_im.paste(ims[i], (x, y))
            i += 1
            y += thumbnail_height
        x += thumbnail_width
        y = 0

    new_im.save(f"256_{str(samples)}_{iteration}.jpg")
```

```python
import os

# directory_path = f"log_256/{directory}/images"
directory_path = img_dir
image_list = []

for filename in os.listdir(directory_path):
    file_path = os.path.join(directory_path, filename)
    if os.path.isfile(file_path):
        image_list.append(f"{directory_path}/{filename}")
```

```python
create_collage(6000, 6000, image_list)
```
