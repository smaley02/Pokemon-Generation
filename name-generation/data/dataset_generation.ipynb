# extract all 1025 pokemon names from https://bulbapedia.bulbagarden.net/wiki/List_of_Pok%C3%A9mon_by_name

```python
# ([A-Z])+([a-z])+([- ]?)(([A-Z])+([a-z])+)?
```

```python
import re

input_file = "raw_pokemon.txt"
output_file = "./input/all_pokemon.txt"

with open(input_file, 'r') as f:
    text = f.read()
```

```python
pattern = re.compile(r'[A-Z][a-z]+(?:[. -][A-Z][a-z]+)?')

matches = pattern.findall(text)
formatted_matches = ["".join(match) for match in matches]

formatted_matches
```

```python
formatted_matches = set(formatted_matches)
```

```python
formatted_matches.remove("Pok")
formatted_matches.remove("No")
formatted_matches.remove("Glitch")
```

```python
len(formatted_matches) # matches the 1025 existing
```

```python
formatted_matches = list(formatted_matches)
formatted_matches = [name.lower() for name in formatted_matches]
```

```python
with open(output_file, 'w') as outfile:
    outfile.write("\n".join(formatted_matches))
```

# extract all 1444 digimon names from https://wikimon.net/Visual_List_of_Digimon

```python
import re

input_file = "raw_digimon.txt"
output_file = "./input/all_digimon.txt"

with open(input_file, 'r') as f:
    text = f.read()
```

```python
remove_pattern = r"\([^)]*\)"
text = re.sub(remove_pattern, "", text)
```

```python
text = text.split("\n")
```

```python
i = 0
n = len(text)
while i < n:
    if not re.search(r'mon', text[i]):
        text.pop(i)
        n -= 1
    else:
        i += 1
```

```python
text = list(set(text))
```

```python
formatted_text = [name.lower() for name in text]
len(formatted_text)
```

```python
with open(output_file, 'w') as outfile:
    outfile.write("\n".join(formatted_text))
```

# concatenated dataset

```python
pokemon = formatted_matches
digimon = formatted_text

all_names = pokemon + digimon
```

```python
with open("./input/pokemon_and_digimon.txt", 'w') as outfile:
    outfile.write("\n".join(all_names))
```
